```python
def reverse_string(my_string):
    yield my_string[::-1]

for c in reverse_string("Elzero"):
    print(c)
```

```python
def decorator(func):
    def wrapper():
        print("sugar added")
        func()
        print("=" * 20)
    return wrapper

@decorator
def my_function2():
    print("tea !!")

@decorator
def my_function():
    print("coffee !!")

my_function2()
my_function()
```
